```python
import pandas as pd
from pyproj import Proj, transform
```

```python
long_lat_buurt = pd.read_csv('../data/longlatbuurt.csv', encoding = "ISO-8859-1")

predictions = pd.read_csv('../data/predictions/predictions_2016.csv', encoding = "ISO-8859-1")
long_lat_buurt = pd.read_csv('../data/longlatbuurt.csv', encoding = "ISO-8859-1")
predictions = predictions.rename(columns={'WijkenEnBuurten':'buurt'})
predictions_with_coords = predictions.merge(long_lat_buurt, on='buurt', how='left')
predictions_with_coords = predictions_with_coords.rename(columns={'long':'X'})
predictions_with_coords = predictions_with_coords.rename(columns={'lat':'Y'})
for year in range(17,30):
    predictions = pd.read_csv('../data/predictions/predictions_20' + str(year) + '.csv', encoding = "ISO-8859-1")
    predictions = predictions.rename(columns={'WijkenEnBuurten':'buurt'})
    predictions_with_coords_temp = predictions.merge(long_lat_buurt, on='buurt', how='left')
    predictions_with_coords_temp = predictions_with_coords_temp.rename(columns={'long':'X'})
    predictions_with_coords_temp = predictions_with_coords_temp.rename(columns={'lat':'Y'})
    predictions_with_coords = pd.concat([predictions_with_coords, predictions_with_coords_temp], axis=0, ignore_index=True)
    
epsgCoordsrow = []
for _, row in predictions_with_coords.iterrows():
    epsgCoords = transform(Proj(init='epsg:3857'), Proj(init='epsg:4326'), row['X'], row['Y'])    
    epsgCoordsrow.append(list(reversed(epsgCoords)))
predictions_with_coords['lat'] = list(map(lambda x: x[0], epsgCoordsrow))
predictions_with_coords['long'] = list(map(lambda x: x[1], epsgCoordsrow))

predictions_with_coords['year'] = predictions_with_coords['year'].astype(str) + '-01-01T00:00Z'
predictions_with_coords
# predictions_with_coords.to_csv('predictions_2016-2029_with_long_lat_coords.csv')
    
```
